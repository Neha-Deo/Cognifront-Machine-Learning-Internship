**1] Importing Library & Acquire the Data**

```python
import pandas as pd

df = pd.read_csv("/content/employee_attrition.csv") 
```

**2] Preprocess the Data**

```python
# Calculating all the missing values for all the columns

df.isna().sum()
```

```python
# Gives Statistical summary of the dataset

df.describe()
```

As we can see the value of min, max and median (50%) of each columns are varying from each other. So Applying Standard Scaler later to normalize the data. With this 2nd Phase of Machine Learning i.e. Data Preprocessing starts.

From the above descirbe table, we can take decision of removing the column "Over 18" because all the above employees are 18+. Since the column Over18 becomes meaning less.

Also Employee count, Employee number can be omitted. Employee count is one for every employees and employee number won't create any impact on attrition rate.

```python
df.drop(['EmployeeCount','EmployeeNumber','Over18','StandardHours'], axis = 1, inplace = True)
```

```python
df.head(3)
```

As we can see from the dataset head(), there are certain columns that are categorical columns which needs to be converted to the numerical ones before training our model. In order to do that we are going to use Label Encoder.

```python
from sklearn.preprocessing import LabelEncoder
```

```python
var_mod = df.columns
le = LabelEncoder()
for i in var_mod:
    df[i] = le.fit_transform(df[i])
```

```python
df.head(7)
```

Every data set has input as well as output variables. 

In this Case Study, Age, BusinessTravel, DailyRate, Department, DistanceFromHome, Education, EducationField, EnvironmentSatisfaction, Gender,PerformanceRating, RelationshipSatisfaction, etc are our input variables. And “Attrition” is our output variable. So let's separate our dataframe df into X and y. 

X will store the input variable & y will store the output variable.

```python
X = df.drop("Attrition",axis=1)
y = df["Attrition"]

# X will drop Attrition and contains all the remaining inputs.
# y will contains Attrition.
```

Now, to proceed further we have to divide our data into the training part and testing part. 

Training part will be used to train our model. The testing part will be used to test our model whether it gives correct predictions or not & also to measure the model's performance. 

So divide X and y into X_train,X_test and y_train,y_test respectively.

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)

# test size is 0.2 that is 20% so train size will be 80% automatically.
# random state is 112 or any numeric number. 
# Every random state number has a different fashion to select data rows randomly.
# random_state simply sets seed to the random generator, so that your train-test splits are always deterministic.
```

```python
# As our data is not normally distributed, we will apply Standard Scaler.

from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
X_train=sc.fit_transform(X_train)
```

```python
X_test=sc.transform(X_test)
```

Now, we have to enter into the 3rd Phase of Machine Learning life cycle. In this phase, we will apply an algorithm to our dataset. So that our model gets trained and able to observe and analyse patterns within data. 

So, lets apply KNN Algorithm in this Case Study.

**3] Train the Model**

```python
from sklearn.neighbors import KNeighborsClassifier
my_model = KNeighborsClassifier(n_neighbors = 3)
result = my_model.fit(X_train,y_train)

"""
my_model is an object of KNN. It will store all the methodology of KNN
Here K=3, where 3 is optimal K which we got through SSQ method.
Now we are applying the KNN methodology to our dataframe by calling a fit method on my_model.
And all the analysis, observations will be stored in the result variable.
So, the result variable contains all the knowledge of dataset which our machine have got through the training phase.

"""
```

The 4th Phase of Machine Learning life cycle is Testing. We will take a test of our model and find out what predictions it gives. 

We can see what attrition status our model predicts for specific input attributes.

**4] Test the Model**

```python
# We are applying predict method on result variable and passing X_test to get predictions.

predictions = result.predict(X_test)
```

```python
predictions

# We get predictions as follows
```

Now, we will Measure our model's performance by using Performace Matrix

```python
print('With KNN (K=3) accuracy is: ', result.score(X_test,y_test))
```

The last step of the Machine Learning life cycle is Deployment. In this, we will give a new entry of input variables and get predicted output i.e. Attrition status by model.
