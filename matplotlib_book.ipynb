**Import book.csv file**

```python
df=pd.read_csv("/content/book.csv")
```

```python
print(df)
```

**Splitting Feature and Target Variable**

```python
X=df.drop('Height',axis=1)
y=df['Height']
```

```python
plt.plot(X,y)
plt.show
```

```python
plt.title("First Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.xticks(np.arange(40,100,10))
plt.yticks(np.arange(130,200,10))
plt.plot(X,y)
plt.show
```

```python
plt.plot(X,y,"g^")
plt.title("First Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show
```

```python
a = np.arange(1,5)
b = a**3
plt.plot(a,b,'r^')
plt.title("General Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show
```

```python
plt.plot(X,y,"go")
plt.title("Second Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show
```

**BAR GRAPHS**

```python
weight = ["40","50","60","70"]
height = [140,150,160,190]
plt.bar(weight,height,color="blue")
plt.title("Bar Graph")
plt.xlabel("Weight")
plt.ylabel("Height")
plt.show()
```

```python
weight = ["40","50","60","70"]
height = [140,150,160,190]
Explode=[0,0.1,0,0]
plt.pie(height, explode=Explode, labels=weight,shadow=True,startangle=60)
plt.axis('equal')
plt.legend(title='List of Firms')
plt.show()
```

**HISTOGRAMS**

```python
x=np.random.randn(1000)
plt.title("Histogram")
plt.xlabel("RandomData")
plt.ylabel("Freqency")
plt.hist(x,10)
plt.show()
```
