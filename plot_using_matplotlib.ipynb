**1] Plot the simple line chart of following data:**

**x = [1, 5, 8, 9, 7, 11, 8, 12, 14, 9, 5]**

**y = [3, 7, 9, 6, 4, 5, 14, 7, 6, 16, 12]**

```python
import matplotlib.pyplot as plt
import numpy as np

plt.xticks(np.arange(0, 51, 1))
plt.yticks(np.arange(0, 20, 1))
plt.title("Simple Line Chart")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

x = [1, 5, 8, 9, 7, 11, 8, 12, 14, 9, 5] 
y = [3, 7, 9, 6, 4, 5, 14, 7, 6, 16, 12]

plt.plot(x,y)
plt.show
```

**2] Create a Pie Chart with a title popularity of Languages:**

**Languages = [hindi, marathi, tamil, telgu, english, gujrathi]**

**Popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]**

```python
import matplotlib.pyplot as plt
import numpy as np

Languages = ["hindi", "marathi", "tamil", "telgu", "english", "gujrathi"]
Popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
Explode = [0, 0, 0, 0.1, 0, 0]

plt.pie(Popularity, explode=Explode, labels=Languages, shadow=True, startangle=30)
plt.axis('equal')
plt.title("Popularity of Languages")
plt.show
```

**3] Plot Bar Chart of following data:**

**Name = [Rudra, Mansvi, Jagruti, Neha, kiran]**

**Percentage = [71, 60, 85, 56, 90]**

```python
import matplotlib.pyplot as plt
import numpy as np

Name = ["Rudra", "Mansvi", "Jagruti", "Neha", "kiran"]
Percentage = [71, 60, 85, 56, 90]
plt.bar(Name, Percentage, color="Blue")
plt.xlabel("Name")
plt.ylabel("Percentage")
plt.title("Bar Chart")
plt.show
```
