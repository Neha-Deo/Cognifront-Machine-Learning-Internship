**Objective : In this Case Study, the target variable is Type. We have to split the data into training and testing parts i.e. 80% for training and 20% for testing. Note down the output of the program for the following :

```python
import pandas as pd
df = pd.read_csv("/content/avocado.csv")
df
```

```python
df.shape
```

```python
df=df.drop(["Unnamed: 0","Date"],axis=1)
df
```

```python
df.describe()
```

```python
df['type'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
var_mod = ['type','year','region']
le = LabelEncoder()
for i in var_mod:
    df[i] = le.fit_transform(df[i])
df
```

```python
X = df.drop(['type'], axis = 1)
y = df['type']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler 
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
from sklearn.svm import SVC
my_model = SVC(kernel = 'rbf', random_state = 0)
result = my_model.fit(X_train, y_train)
```

```python
predictions = result.predict(X_test)
predictions
```

```python
from sklearn import metrics
print("\n Accuracy : ", metrics.accuracy_score(y_test, predictions))
```

```python
import seaborn as sn
from sklearn.metrics import confusion_matrix
conf_matrix =confusion_matrix(predictions,y_test)
confusion_df = pd.DataFrame(conf_matrix, index=['Actual 0','Actual 1'], columns=['Predicted 0','Predicted 1'])
confusion_df
```
